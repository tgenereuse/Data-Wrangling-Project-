### 6. Write a function that takes each value of the "Score" column and returns "happy" if the Score value is greater than 6 and "not-happy" if the score value is less than or equal to 6. Apply the function to the "Score" column, create a new column from it, and call it "Happy- notHappy".

```python
def happiness_level(Column):
    Score = Column[3]
    if Score <= 6:
        return('Not_Happy') 
    else:
        return ('Happy')


World_happiness["Happy_notHappy"] = World_happiness.apply(happiness_level, axis = "columns")

World_happiness.head(2)
```

```python
# According to the below output countries are generally not happy since they scored almost two times higher as not happy compared to happy 

World_happiness['Happy_notHappy'].value_counts()
```
